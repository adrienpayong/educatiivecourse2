```python
import cv2 as cv
from matplotlib import pyplot as plt
import os


class DepthMapStereo:
    def __init__(self, left_image, right_image):
        self.left_image = left_image
        self.right_image = right_image

    def depth(self):
        # reading both of the images in grayscale format
        imgLeft = cv.imread(self.left_image, 0)
        imgRight = cv.imread(self.right_image, 0)
        """
        numDisparities the disparity search range. For each pixel algorithm will find the best
         blockSize the linear size of the blocks compared by the algorithm. The size should be odd
        """
        stereoimage = cv.StereoBM_create(numDisparities=14, blockSize=13)
        disparity = stereoimage.compute(imgLeft, imgRight)
        plt.imshow(disparity, 'gray')
        plt.show()
```

```python
### 8. Depth Map on Stereo Images ###
left_image = os.path.join("view0.png")
right_image = os.path.join("view2.png")
depth_map_obj=DepthMapStereo(left_image,right_image)
depth_map_obj.depth()
```
